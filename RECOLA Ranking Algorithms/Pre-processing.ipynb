# Pre-processing

```python
import os
import pandas as pd
import shutil
```

```python
project_directory = r'C:\Users\marco\OneDrive\Desktop\Final Year Project'
os.chdir(project_directory)
base_dir = os.getcwd() 
```

```python

def processing_files(source_directory, destination_directory):

    if not os.path.exists(destination_directory):
        os.makedirs(destination_directory)

    for root, _, files in os.walk(source_directory):
        for file in files:
            if file.endswith('.csv'):
                source_file_path = os.path.join(root, file)
                destination_file_path = os.path.join(destination_directory, file)
                shutil.copy2(source_file_path, destination_file_path)

    for file in os.listdir(destination_directory):
        if file.endswith('.csv'):
            file_path = os.path.join(destination_directory, file)
            try:
                df = pd.read_csv(file_path)

                if df.isnull().values.any():
                    print(f"File '{file}' contains missing values. Removing it.")
                    os.remove(file_path)
            except Exception as e:
                print(f"Error processing file '{file}': {e}")

    print("Finished processing files. The directory with processed files has been updated.")

# File paths
input_path = 'Data/RECOLA'
destination_path = 'RECOLA Ranking Algorithms/RECOLA_Processed_Data'

processing_files(input_path, destination_path)
```
